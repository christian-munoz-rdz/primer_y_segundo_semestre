> Realice una función que **reciba** 2 párametros una cadena y un carácter. A partir de estos parámetros la función debe indicar la posición en la cadena en la cual coincide con el caracter buscado y también indicar cuantas veces apareció el carácter en la cadena.

```python
def func(cadena,caracter):
  cadena = list(cadena)
  for i in range(len(cadena)):
      if  cadena[i] == caracter:
        print(f"{caracter} se encuentra en la posición {i}")
  
  print(f"Coincidencias totales: {cadena.count(caracter)}")

cadena = (input("Ingrese una cadena: ")).lower()
caracter = input("Ingrese un caracter: ")
func(cadena,caracter)
```

> Realice una función que **reciba** dos números y **devuelva** el punto intermedio entre ambos números. **Nota:** El número intermedio de dos números corresponde a la suma de los dos números dividido entre 2.

```python
def inter(a,b):
  intermedio = (a+b)/2
  return f"El punto intermedio de {a} y {b} es {intermedio}"

a = int(input("Ingrese un primer número-->"))
b = int(input("Ingrese otro número--->"))

inter(a,b)
```
